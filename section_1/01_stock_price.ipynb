# 株価の取得

```python
# 1. ライブラリのインストール（yfinanceがインストールされていない場合はコメントを外してください）
#!pip install yfinance

# 2. ライブラリのインポート
import yfinance as yf

# 3. データの取得
# 例: トヨタ自動車 (証券コード: 7203)
# 日本株は証券コードの末尾に ".T" をつけるのがルールです
data = yf.download("7203.T", period="1mo", auto_adjust=True) # 過去1ヶ月分

# 4. データの表示
data
```

# 株式の情報

```python
# 1. Tickerオブジェクトを作成（ここではトヨタ自動車: 7203.T）
ticker = yf.Ticker("7203.T")

# 2. 情報を取得 (.info に辞書形式で入っています)
info = ticker.info
info
```

情報を個別に取得することもできます。

```python
print(f"【時価総額】 {info.get('marketCap'):,} 円")
print(f"【PER (株価収益率)】 {info.get('trailingPE')}")
print(f"【PBR (株価純資産倍率)】 {info.get('priceToBook')}")
print(f"【配当利回り】 {info.get('dividendYield', 0) * 100:.2f} %")
```
